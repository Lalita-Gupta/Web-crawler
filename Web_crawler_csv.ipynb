```python
# Web crawler using BeautifulSoup in python.

# frameworks used

import requests;
import bs4;
import csv;
import os;

# clear file data
f = open("Websites.csv", "w"); # csv file to tore all links extracted.
f.truncate();
f.close();

# User has the choice of the website to be crawled and number of links to be extracted from that website.
site = input("Enter the website to be crawled: ") 
n = int(input("Enter number of first unique links to be printed or print until all links get exhausts: "))

res = requests.get(site); # Web crawling used on this website 

soup = bs4.BeautifulSoup(res.text, 'lxml');

list1 = [];

for link in soup.find_all('a', href=True):
    
    # condition given according to this website
    if link['href'][0] == "#":
        pass;
    elif link['href'][0] == "/" and link['href'][1] != "/":
        list1.append(site + link['href']);
    elif link['href'][:2] == "//":
        list1.append('https:'+ link['href']);
    else:
        list1.append(link['href']);
        
print("Total number of websites:", len(list1));    
set1 = set(list1); # for unique links
counter = 1;

# print first n unique links or print until all links get exhausts.
for i in set1:
    if counter <= n:
        print(counter, ")", i);
        with open('Websites.csv', 'a', newline = '') as file:
            writer = csv.writer(file);
            writer.writerow([counter, ")", i]); # writing links in csv file.
    else:
        break;
    counter += 1;
file.close(); # csv file closed.
```
